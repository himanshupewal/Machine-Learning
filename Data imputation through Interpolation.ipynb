```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### LINEAR INTERPOLTION
```

```python
df1 = np.array([1,2,3,4,5])
df2 = np.array([2,4,6,8,10])
```

```python
plt.plot(df1,df2)
```

```python
plt.scatter(df1,df2)
```

```python
new_x = np.linspace(1,5,10)
y_interpol = np.interp(new_x,df1,df2)
print(y_interpol)
```

```python
plt.scatter(new_x,y_interpol)
```

```python
## CUBIC INTERPOLATION
```

```python
x = np.array([1,2,3,4,5])
y= np.array([1,4,9,16,25])
```

```python
from scipy.interpolate import interp1d
```

```python
f = interp1d(x,y,kind="cubic")
```

```python
new_x = np.linspace(1,5,10)
y_interp = f(new_x)
print(y_interp)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(new_x,y_interp)
```

```python
## polynomial
```

```python
x = np.array([1,2,3,4,5])
y= np.array([1,4,9,16,25])
```

```python
p = np.polyfit(x,y,2)
```

```python
new_x= np.linspace(1,5,10)
y_inter = np.polyval(p,new_x)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(new_x,y_inter)
```
