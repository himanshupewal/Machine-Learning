```python
# SMOTE - SYNTHATIC MINORITY OVERSAMPLING TECHINQUES
#  ANOTHER  TECHINQUE TO HANDEL IMBALANCE DATA 
# INTERPOLATION
```

```python
from sklearn.datasets import make_classification 
```

```python
X,Y = make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=1)
```

```python
X
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(X,columns=["f1","f2"])
```

```python
df2 = pd.DataFrame(Y,columns=["target"])
```

```python
df2
```

```python
data = pd.concat([df1,df2],axis=1)
```

```python
data["target"].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data["f1"],data["f2"],c=data["target"])
```

```python
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE 
```

```python
oversampling = SMOTE()
x,y = oversampling.fit_resample(data[["f1","f2"]],data["target"])
```

```python
x.shape
```

```python
y.shape
```

```python
df1 = pd.DataFrame(x,columns=["f1","f2"])
df2 = pd.DataFrame(y,columns=["target"])
oversampling_data = pd.concat([df1,df2],axis=1)
plt.scatter(oversampling_data["f1"],oversampling_data["f2"],c=oversampling_data["target"])
```
